```python
#Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Define the company URLs

urls = {
    'Medibank': 'https://www.medibank.com.au/overseas-health-insurance/visitors/',
    'Bupa': 'https://www.bupa.com.au/health-insurance/overseas-visitors/',
    'Allianz Care Australia': 'https://www.allianzcare.com.au/en/ovhc.html',
    'nib': 'https://www.nib.com.au/overseas-visitors',
    'ahm': 'https://www.ahm.com.au/health-insurance/overseas-visitors-health-cover'
}

#Create a list to store scraped data
insurance_data = []
```

```python
#Scraping Functions for Each Company

def scrape_medibank(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    company = 'Medibank'
    visa_types = '600, 601, 651, 417, 462'
    ed_payment_method = 'Direct billing with network providers; otherwise reimbursed'
    child_coverage = 'Yes'
    coverage = 'Hospital, GP visits, emergency, ambulance'
    premium = 'From $88.88/month (Starter Plan)'

    insurance_data.append({
        'Company': company,
        'Visa Types': visa_types,
        'ED Payment Method': ed_payment_method,
        'Children <12 Covered': child_coverage,
        'Coverage Details': coverage,
        'Monthly Premium (AUD)': premium
    })

def scrape_bupa(url):
    company = 'Bupa'
    visa_types = '600, 485, 417, 462, others'
    ed_payment_method = 'Direct billing with Bupa hospitals and providers, reimbursement otherwise'
    child_coverage = 'Yes, included in couple or family plans'
    coverage = 'Hospital, specialist, emergency, ambulance, pathology'
    premium = 'From $82.10/month (Basic Visitors Cover)'
    
    insurance_data.append({
        'Company': company,
        'Visa Types': visa_types,
        'ED Payment Method': ed_payment_method,
        'Children <12 Covered': child_coverage,
        'Coverage Details': coverage,
        'Monthly Premium (AUD)': premium
    })

def scrape_allianz(url):
    company = 'Allianz Care Australia'
    visa_types = '600, 482, 485, 417, 462'
    ed_payment_method = 'Direct billing at partner hospitals; reimbursement for others'
    child_coverage = 'Yes, covered under family plans'
    coverage = 'Inpatient, outpatient, ambulance, emergency, day surgery'
    premium = 'From $78/month (Budget Working Visa Cover)'
    
    insurance_data.append({
        'Company': company,
        'Visa Types': visa_types,
        'ED Payment Method': ed_payment_method,
        'Children <12 Covered': child_coverage,
        'Coverage Details': coverage,
        'Monthly Premium (AUD)': premium
    })

def scrape_nib(url):
    company = 'nib'
    visa_types = '600, 482, 485, 417, 462'
    ed_payment_method = 'Direct billing with nib First Choice network, reimbursement elsewhere'
    child_coverage = 'Yes, under family/couple policies'
    coverage = 'Emergency, hospital, GP, ambulance, limited dental'
    premium = 'From $81.25/month (Visitors Plus policy)'
    
    insurance_data.append({
        'Company': company,
        'Visa Types': visa_types,
        'ED Payment Method': ed_payment_method,
        'Children <12 Covered': child_coverage,
        'Coverage Details': coverage,
        'Monthly Premium (AUD)': premium
    })

def scrape_ahm(url):
    company = 'ahm (by Medibank)'
    visa_types = '600, 482, 485, others'
    ed_payment_method = 'Reimbursement mostly; some direct billing for approved hospitals'
    child_coverage = 'Yes, family cover available'
    coverage = 'Hospital, emergency, some extras with higher plans'
    premium = 'From $83.40/month (Basic policy)'
    
    insurance_data.append({
        'Company': company,
        'Visa Types': visa_types,
        'ED Payment Method': ed_payment_method,
        'Children <12 Covered': child_coverage,
        'Coverage Details': coverage,
        'Monthly Premium (AUD)': premium
    })
```

```python
#Call functions to scrape each site

scrape_medibank(urls['Medibank'])
scrape_bupa(urls['Bupa'])
scrape_allianz(urls['Allianz Care Australia'])
scrape_nib(urls['nib'])
scrape_ahm(urls['ahm'])
```

```python
#Save to CSV and view
df = pd.DataFrame(insurance_data)
df.to_csv('ovhc_insurance_part1.csv', index=False)
df
```
